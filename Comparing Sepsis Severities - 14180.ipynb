# Simulating MIMIC Patient
We used Pulse to simulate patients in MIMIC, initializing them using their readings in MIMIC, and administering IV fluids and vasopressors at each time step as done in MIMIC. (See [MIMICSimulator.py](https://gitlab.kitware.com/faaizT/engine/-/blob/local-changes/src/python/pulse/rlengine/SimulationEngine.py) for more details about the simulations.)  
This note compares a patient simulated in Pulse with different sepsis severities.  

```python
import pandas as pd
icustayid = 14180.0
export_dir = "/data/localhost/taufiq/export-dir/"
mimic_csv = "/data/localhost/taufiq/Pulse/engine/src/python/pulse/rlengine/resources/MIMICtable-1hourly_entire-stay.csv"
```

```python
simulation_0 = pd.read_csv(export_dir + "simulation_sepsis_0.0_" + str(icustayid) + "_final_1613392976.csv")
simulation_0 = simulation_0.replace(r"[\[|\]]", "", regex=True)
```

```python
simulation_0_25 = pd.read_csv(export_dir + "simulation_sepsis_0.25_" + str(icustayid) + "_final_1613403192.csv")
simulation_0_25 = simulation_0_25.replace(r"[\[|\]]", "", regex=True)
```

```python
simulation_0_5 = pd.read_csv(export_dir + "simulation_sepsis_0.5_" + str(icustayid) + "_final_1613418352.csv")
simulation_0_5 = simulation_0_5.replace(r"[\[|\]]", "", regex=True)
simulation_0_75 = pd.read_csv(export_dir + "simulation_sepsis_0.75_" + str(icustayid) + "_final_1613433561.csv")
simulation_0_75 = simulation_0_75.replace(r"[\[|\]]", "", regex=True)
simulation_1 = pd.read_csv(export_dir + "simulation_sepsis_1.0_" + str(icustayid) + "_final_1613448614.csv")
simulation_1 = simulation_1.replace(r"[\[|\]]", "", regex=True)
```

```python
time = simulation_0['time']
diaBP_0 = simulation_0['DiastolicArterialPressure (mmHg)']
diaBP_0_25 = simulation_0_25['DiastolicArterialPressure (mmHg)']
diaBP_0_5 = simulation_0_5['DiastolicArterialPressure (mmHg)']
diaBP_0_75 = simulation_0_75['DiastolicArterialPressure (mmHg)']
diaBP_1 = simulation_1['DiastolicArterialPressure (mmHg)']
```

```python
import matplotlib.pyplot as plt
plt.plot(time, pd.to_numeric(diaBP_0))
plt.plot(time, pd.to_numeric(diaBP_0_25))
plt.plot(time, pd.to_numeric(diaBP_0_5))
plt.plot(time, pd.to_numeric(diaBP_0_75))
plt.plot(time, pd.to_numeric(diaBP_1))
plt.xlabel('Time(s)')
plt.ylabel('mmHg')
plt.title('Diastolic Arterial Pressure')
plt.legend(['Severity: 0', 'Severity: 0.25', 'Severity: 0.5', 'Severity: 0.75', 'Severity: 1.0'])
plt.show()
```
